# Spotify Business Intelligence Project - Day 2

Step 2 : Data Cleansing

Organize the data so it tells a story. The Spotify exec wants to know about the rhythms of success – think time !

Mission 2.1 : Create a "Date_str" column using released_year, released_month and released_day

Tips: check how to concatenate columns in dataframes

```python
df["Date_str"] = df['released_year'].astype(str) +"/"+ df["released_month"].astype(str) +"/"+ df["released_day"].astype(str)
df["Date_str"]
```

Mission 2.2 :  Display the type of the dataframe columns

```python
df.dtypes
```

Notice the last created column has not a date type

Mission 2.3 : Create a new Column "Date" that contains the same data as the columns "Date_str" but convert the data into a datetime type and display the dataframe columns data type again

```python
date=pd.to_datetime(df['Date_str'], format="%Y/%m/%d")
df['Date']=date
df['Date']
df.dtypes
```

Mission 2.4 : Create a new dataframe named 'sorted_df', but ensure that the dataframe is sorted in ascending order based on the 'Date_str' column.

```python
sorted_df = df.sort_values(by=['Date'], ascending=True)
sorted_df
```

Mission 2.5 : Create a new columns 'Quarter' based on the columns 'Date'

```python
df['Quarter'] = pd.PeriodIndex(df.Date, freq='Q')
df.head()
```

# Spotify Business Intelligence Project - Day 4

Step 4 : Visualization

Now, it's time to create visualizations – the Spotify executive loves pictures. Build dashboards that showcase your insights. Think charts, graphs, and cool visuals!

Mission 4.1 : How do the streaming counts vary across different quarters of the year?

```python
plt.figure(figsize=(12, 6))

# Aggregate total streams for each quarter
quarterly_streams = sorted_df.groupby('Quarter')['streams'].sum().reset_index()

# Bar plot
sns.barplot(x='Quarter', y='streams', data=quarterly_streams, palette='muted')
plt.title('Total Streaming Counts Across Quarters')
plt.xlabel('Quarter of the Year')
plt.ylabel('Total Streams')
plt.show()
```

Note: The requested projection was regarding the stream counts vs the year but the date dimension available within the dataset is related to the release date so the previous chart is not answering the question, in fact being able to understand the meaning of the data is an important part of the analysis

Mission 4.2 : What is the distribution of danceability percentages in the dataset?

```python
sns.distplot(sorted_df['danceability_%'], bins=20, kde=True)
plt.title('Distribution of Danceability')
plt.xlabel('Danceability (%)')
plt.ylabel('Frequency')
plt.show()
```

Mission 4.3 : What is the distribution of songs across different keys in terms of the number of streams?

```python
plt.figure(figsize=(8, 8))

# Aggregate total streams for each key
key_streams = sorted_df.groupby('key')['streams'].sum().reset_index()

# Pie chart
plt.pie(key_streams['streams'], labels=key_streams['key'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette('viridis'))
plt.title('Distribution of Streams Across Different Keys')
plt.show()
```

Mission 4.4 : What is the distribution of song tempo (BPM) and how does it relate to their danceability?

```python
sns.jointplot(data=sorted_df, x='bpm', y='danceability_%', kind='scatter')
plt.suptitle('Distribution of BPM and Danceability')
plt.show()
```
